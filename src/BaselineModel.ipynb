```python
import pandas as pd
import numpy as np
import lgAimersDPP as dpp

from sklearn.ensemble import ExtraTreesClassifier
```

```python
pd.set_option('display.max_columns',None) # 전체 열 출력하기
dpp
```

```python
train = pd.read_csv('data/train.csv')
train = dpp.dpp(train)
print(train.shape)
train.head()
```

```python
pd.set_option('display.max_rows',None) # 전체 행 출력하기
print(train.isnull().sum())
pd.set_option('display.max_rows',10)
```

```python
test = pd.read_csv('data/test.csv')
```

```python
X_train = train.drop('임신 성공 여부', axis=1)
Y_train = train['임신 성공 여부']
model = ExtraTreesClassifier(random_state=42)

model.fit(X_train, Y_train)
```

```python
model.feature_importances_
```

```python
test.head()
```

```python
#test 데이터 전처리
test = dpp.dpp(test)
print(test.shape)
test.head()
```

```python
pred_proba = model.predict_proba(test)[:,1]
```

```python
pred_proba
```

```python
res = pd.read_csv('data/sample_submission.csv')
res['probability'] = pred_proba
```

```python
res.to_csv('data/0214baseline_submit1.csv', index=False)
```
