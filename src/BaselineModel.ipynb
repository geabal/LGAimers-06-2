```python
test_di_id = test_di['ID']
test_ivf_id = test_ivf['ID']
```

```python
X_test_di = test_di.drop(['ID','시술 시기 코드','시술 유형'], axis=1)
X_test_ivf = test_ivf.drop(['ID','시술 시기 코드','시술 유형'], axis=1)

pred_proba_di = model_di.predict_proba(X_test_di)[:,1]
pred_proba_ivf = model_ivf.predict_proba(X_test_ivf)[:,1]
```

```python

#ID 기준으로 test 데이터 합치기. ID, 임신 성공 여부 컬럼만 반환
def concatRes(di, ivf, di_id, ivf_id):
    lenres = len(di) + len(ivf)
    print(lenres)
    maxdi = len(di)
    maxivf = len(ivf)
    rows = []
    idxd, idxi = 0,0
    for i in range(lenres):
        numi = int(ivf_id[idxi][-5:])
        numd = int(ivf_id[idxd][-5:])
        if i == numi:
            rows.append([ivf_id.iloc[idxi], ivf[idxi]])
            idxi += 1
            if idxi == maxivf:
                idxi -= 1
        else:
            rows.append([di_id.iloc[idxd], di[idxd]])
            idxd += 1
            if idxd == maxdi:
                idxd -= 1
    res = pd.DataFrame(rows, columns = ['ID', 'probability'])
    return res
```

```python
res = concatRes(pred_proba_di, pred_proba_ivf, test_di_id , test_ivf_id)
res
```
