- 오버 피팅 문제 해결을 위해 cross validation 수행  

## 결측치 처리  
  
  
1. 반복적으로 나타나는 일정한 결측치 수  
  
결측치 수가 동일한 컬럼이 다수 존재한다.(해당 행의 수 6291)  
  
- '시술 유형'이 DI인 경우, IVF와 관련된 29개의 column이 결측치로 기록됐다.  
- 결측치 처리 방향: 시술 유형이 'DI'인 경우와 'IVF'인 경우를 나누어 모델링(데이터 분리) 
  
2. PGS, PGD검사 결측치  
  
25만 6천개의 데이터 중 25만개 이상 결측치 확인.  
PGD, PGS는 시험관 시술을 수행할 때만 선택적으로 수행하는 것.  
  
- IVF시술을 하지 않은 경우 시험관 시술과 관련된 검사인 PGS, PGD검사는 수행하지 않는다. > NaN으로 기록됨  
- '착상 전 유전 검사 사용 여부(PGS)'의 결측치 수가 '착상 전 유전 진단 사용 여부(PGD)'보다 압도적으로 많은 이유는 PGS검사가 이전에 반복적인 유산, 착상 실패 경험이 누적된 경우 시행하는 검사이기 때문에 검사 권유 케이스가 드물기 때문일 수 있다.  
- 결측치 처리 방향: (1) DI 시술을 수행한 경우, 해당 컬럼 삭제 (2) IVF 시술을 수행한 경우도 '착상 전 유전 검사 사용 여부(PGS)' 결측치를 모두 0으로 처리.  
- 이유: 반복적 착상 실패 및 부모의 염색체 비정상 소견을 받은 뒤에 PGS를 진행하기 때문에, PGS의 시행 케이스는 드물다. 또한 검사를 받았는데 NaN으로 기록하는 경우보다는 검사를 고려할 필요가 없어 NaN으로 기록하는 것이 일반적일 것이라 추측됨.    

참고 1: PGS, PGD 차이: https://chaimc.m.chamc.co.kr:447/clinic/gene/introduction.cha  
참고 2: 반복착상실패 검사(PGS) https://m.blog.naver.com/lynn_2020/222399467344  
 

3. 시술 당시 나이 == '알 수 없음'  
  
환자가 시술 당시 자신의 나이를 데이터로 사용하지 않도록 정보를 제공하지 않은 것으로 추측됨.  
환자의 나이를 알 수 없으므로 본인 제공 난자를 사용하는 경우, 난자 제공자의 나이도 알 수 없다.  
- 결측치 처리 방향: 나이를 순서형 데이터로 변환할 때 가장 나이가 많은 범주로 포함킬 것.(6으로 대치)  
- 이유: 시술 당시 나이가 '알 수 없음'인 경우 임신 성공 케이스가 없음. 또한 모델을 낙관적으로 만들기보다는 비관적으로 만들 때 환자가 보다 모델의 예측 결과를 신뢰할 수 있음.  

4. 임신 시도 또는 마지막 임신 경과 연수  
  
전체 25만 6천개의 데이터 중 약 24만 개의 결측치 확인.  
개인 정보 문제로 수집한 데이터보다 수집되지 않은 데이터가 많은 것으로 추측됨.  
  
- 임신 성공 여부와 관계 없이 '임신 시도 또는 마지막 임신 경과 연수' 분포의 형태가 동일하다.  
- 일반적으로 1년 이내로 자연 임신 시도에 실패했을 때 난임 판정을 받게 됨. > 모든 수치를 0으로 대치하거나 최빈값으로 채우게 된다면 데이터에 왜곡이 발생할 가능성이 높음.  
- 결측치 처리 방향: KNN 보간법을 활용해 가장 유사한 데이터들 간에 비슷한 수치를 가지도록 보간  
  
참고: KNN 보간법 https://excelsu.tistory.com/12  
  
5. '~ 경과일' 컬럼  
  
각 시술 후 경과일을 기록한 컬럼으로 결측치가 4만~25만 개 범위로 존재.  
시술을 수행하지 않은 사람들의 데이터가 NaN으로 기록되었다.  
- 각 컬럼은 임신 성공 여부와 관계 없이 분포가 거의 동일하다. > 정확한 경과일보다는 시술 여부를 기록하는 것이 필요
- 결측치 처리 방향: '~ 경과일' 컬럼을 제거하고 '~ 시술 여부' 컬럼을 추가한다. 시술 여부 컬럼은 '~ 경과일' 컬럼이 NaN인 경우 0, 그렇지 않은 경우 1을 기록한다.
